## Workflow
A workflow is a sequence of steps that are performed to complete a task or achieve a goal.

Think of it as a process diagram showing:

    Trigger → Actions → Result
These are three common workflow patterns in n8n and automation systems.

## 1. Linear Workflow

A linear workflow executes steps one after another in a fixed sequence.

```text
Trigger
   ↓
Step A
   ↓
Step B
   ↓
Step C
```

### Example

When a customer submits a form:

```text
Form Submission
      ↓
Save to Database
      ↓
Send Welcome Email
      ↓
Notify Sales Team
```

Each step waits for the previous step to finish.

### Characteristics

* Simple and predictable
* No branching
* Best for straightforward processes

---

## 2. Parallel Workflow

A parallel workflow executes multiple branches simultaneously after a certain step.

```text
         ┌──→ Task A
Trigger ─┤
         ├──→ Task B
         │
         └──→ Task C
```

### Example

When a new order is placed:

```text
New Order
    │
    ├──→ Send Confirmation Email
    │
    ├──→ Update Inventory
    │
    └──→ Notify Warehouse
```

These tasks do not depend on each other, so they can run at the same time.

### Benefits

* Faster execution
* Efficient use of resources
* Independent tasks run simultaneously

### In n8n

You can connect one node to multiple downstream nodes.

```text
Order Trigger
      │
      ├──→ Email Node
      ├──→ Database Node
      └──→ Slack Node
```

---

## 3. Conditional Branching Workflow

A conditional workflow chooses different paths based on rules or conditions.

```text
             Yes ──→ Action A
            /
Trigger → Condition
            \
             No ──→ Action B
```

### Example

Checking whether a customer is premium:

```text
New Customer
      ↓
Is Premium Customer?
      │
      ├── Yes → Assign Priority Support
      │
      └── No  → Assign Standard Support
```

### Another Example

Loan Approval:

```text
Application
      ↓
Credit Score > 700?
      │
      ├── Yes → Approve Loan
      │
      └── No  → Reject Loan
```

### In n8n

This is usually done with:

* IF Node
* Switch Node
* Filter Conditions

---

## Combining All Three

Real-world workflows often use all three patterns.

```text
New Order
     ↓
Check Order Value
     │
     ├── Value > $500
     │         │
     │         ├──→ Notify Manager
     │         └──→ Fraud Check
     │
     └── Value ≤ $500
               │
               ├──→ Send Email
               └──→ Update Inventory
```

Here:

* The process starts **linearly**.
* A **conditional branch** decides which path to take.
* Some tasks run **in parallel** within each branch.

This combination is the foundation of most business and AI automation workflows built in **n8n**.


## Nodes

In **n8n**, a **node** is the smallest functional unit of a workflow. Each node performs one specific operation on data.

You can think of a node as a **single task or function**.

### Types of Nodes

1. **Trigger Nodes**

   * Start a workflow
   * Example: Webhook, Schedule, Email trigger

2. **Action Nodes**

   * Perform operations
   * Example: HTTP Request, Send Email, Insert into Database

3. **Transform Nodes**

   * Modify data
   * Example: Set, IF, Switch, Merge, Code node

### Example

```text id="n1"
Webhook Trigger → HTTP Request → Send Email
```

Each block is a node performing a distinct role.

### Key Idea

A node:

* Takes input data (JSON)
* Processes it
* Produces output data

---

## Connections

A **connection** defines how data flows between nodes.

It is the **link (arrow)** that passes output from one node to another.

### Basic Structure

```text id="c1"
Node A ───→ Node B ───→ Node C
```

### What connections do:

* Define execution order
* Pass data forward
* Control workflow structure

---

## Example Workflow

```text id="ex1"
Form Trigger
     ↓
Set Node
     ↓
IF Node
   ├── Yes → Send Email
   └── No  → Store in Database
```

### Explanation:

* Each arrow is a **connection**
* Data moves along connections
* Branching connections create different paths

---

## Important Concept

### Nodes = What happens

### Connections = When and where it flows

---

## Advanced Insight (Important for n8n)

1. A single node can have **multiple outgoing connections**

   * Enables parallel execution

2. A node can receive input from **multiple nodes**

   * Used in merge operations

3. Execution depends on connections, not placement

   * Layout is visual only

---

## Simple Mental Model

Think of it like a pipeline:

* **Nodes = machines**
* **Connections = conveyor belts**
* **Data = material moving through system**

---


In **n8n**, nodes are broadly classified into two functional categories: **Triggers** and **Actions**. This distinction defines *how a workflow starts* and *what it does afterward*.

---

## 1. Trigger Nodes

A **trigger node** is the entry point of a workflow. It defines *when* and *how* the workflow starts executing.

### Key property:

* It **initiates execution automatically or on an event**

### Common trigger types:

1. **Webhook Trigger**

   * Starts workflow when an external system sends a request
   * Example: form submission, API call

2. **Schedule Trigger**

   * Runs at fixed intervals (cron-based)
   * Example: every day at 9 AM

3. **Email Trigger**

   * Starts when a new email is received

4. **App-based triggers**

   * Example: new row in Google Sheets, new message in Slack

---

### Example

```text id="t1"
Webhook Trigger
      ↓
Process Data
      ↓
Send Email
```

Here, the workflow starts only when the webhook is called.

---

## 2. Action Nodes

An **action node** performs a task *after the workflow has started*. It does not start workflows on its own.

### Key property:

* It **executes only when triggered by a previous node**

### Common action types:

1. **HTTP Request**

   * Calls external APIs

2. **Send Email**

   * Sends emails via Gmail/SMTP

3. **Database Nodes**

   * Insert, update, query data

4. **Transform Nodes**

   * Set, IF, Switch, Merge, Code

5. **AI Nodes**

   * OpenAI, embeddings, summarization, etc.

---

### Example

```text id="t2"
Schedule Trigger
       ↓
HTTP Request (Get Data)
       ↓
OpenAI (Analyze)
       ↓
Email Node (Send Report)
```

Only the first node is a trigger; the rest are actions.

---

## Core Difference

| Feature    | Trigger Node               | Action Node                |
| ---------- | -------------------------- | -------------------------- |
| Role       | Starts workflow            | Performs tasks             |
| Execution  | Event-based / scheduled    | Depends on previous node   |
| Position   | Always first (entry point) | Anywhere after trigger     |
| Dependency | Independent                | Dependent on upstream data |

---

## Important Insight

A workflow **must have at least one trigger node**, otherwise it cannot start.

Everything else in automation is just:

> Trigger → Actions → Result

---

## Simple Mental Model

* Trigger = “WHEN does it start?”
* Action = “WHAT happens next?”

---


## Variables and Expressions in n8n

In **n8n**, **variables and expressions** are used to dynamically access, manipulate, and reuse data inside workflows.

They are essential because workflows are not static—you constantly work with incoming data.

---

# 1. Variables (Concept)

A **variable** is a piece of data stored during workflow execution that can be reused in later nodes.

In n8n, you do not explicitly declare variables like in Python or JavaScript. Instead, variables are **derived from node outputs**.

### Example idea:

If a node returns:

```json
{
  "name": "Rida",
  "score": 95
}
```

Then:

* `name` and `score` act like variables inside the workflow

---

# 2. Expressions

An **expression** is how you access and manipulate data from previous nodes.

In n8n, expressions are written using:

```text
{{ }}
```

---

## Basic Expression Syntax

```text
{{ $json["key"] }}
```

### Example:

If input is:

```json
{
  "name": "Rida"
}
```

Expression:

```text
{{ $json["name"] }}
```

Output:

```
Rida
```

---

# 3. Accessing Data from Previous Nodes

You can reference other nodes directly:

```text
{{ $node["Node Name"].json["key"] }}
```

### Example:

```text
{{ $node["Webhook"].json["email"] }}
```

This fetches data from a specific node, not just the previous one.

---

# 4. Common Built-in Variables

### 1. `$json`

Current item data

```text
{{ $json["email"] }}
```

---

### 2. `$node`

Access any node output

```text
{{ $node["Set"].json["price"] }}
```

---

### 3. `$items`

Access all items from a node

```text
{{ $items("HTTP Request") }}
```

---

### 4. `$now`

Current timestamp

```text
{{ $now }}
```

---

### 5. `$env`

Environment variables

```text
{{ $env["API_KEY"] }}
```

---

# 5. Using Expressions in Real Cases

## Example 1: Email personalization

```text
Hello {{ $json["name"] }}, your order is confirmed.
```

---

## Example 2: API request body

```json
{
  "email": "{{ $json["email"] }}",
  "message": "Welcome {{ $json["name"] }}"
}
```

---

## Example 3: Conditional logic

```text
{{ $json["score"] > 50 }}
```

Returns:

* true or false

---

# 6. Important Concept

Expressions are evaluated **at runtime**, not design time.

Meaning:

* You don’t store values
* You **pull values dynamically from execution data**

---

# 7. Mental Model

Think of expressions like:

> “Pull data from previous nodes and use it dynamically here.”

* Variables = data available in workflow
* Expressions = way to access and manipulate that data

---

# 8. Common Mistakes

### ❌ Wrong

```text
$json.name
```

### ❌ Wrong

```text
{{ name }}
```

### ✅ Correct

```text
{{ $json["name"] }}
```

## Manual vs Automatic Execution in n8n

In **n8n**, execution refers to how and when a workflow is run. There are two primary modes: **Manual Execution** and **Automatic Execution**.

---

# 1. Manual Execution

Manual execution means you **run the workflow yourself inside the editor**.

### How it works

* You click **“Execute Workflow”**
* n8n runs the workflow immediately
* It uses sample or test data (if available)

### Example use cases

* Testing a new workflow
* Debugging node logic
* Checking API responses
* Validating expressions and JSON structure

### Example flow

```text id="m1"
User Click → Execute Workflow → Run Nodes → Output Displayed
```

### Key characteristics

* Runs on demand
* Requires user interaction
* Used for development/testing
* Often uses mock or test input

---

# 2. Automatic Execution

Automatic execution means the workflow runs **without human intervention**, triggered by an event or schedule.

### Trigger sources

* Webhook request
* Scheduled time (cron)
* New email/message
* Database update
* App event (Google Sheets, Slack, etc.)

### Example flow

```text id="a1"
Event Happens → Trigger Node → Workflow Executes Automatically
```

### Example

```text id="a2"
Every 9 AM (Schedule Trigger)
      ↓
Fetch Data
      ↓
Generate Report
      ↓
Send Email
```

### Key characteristics

* Fully automated
* Runs in production environments
* Event-driven or time-driven
* No user interaction required

---

# 3. Key Differences

| Feature     | Manual Execution       | Automatic Execution       |
| ----------- | ---------------------- | ------------------------- |
| Trigger     | User click             | Event or schedule         |
| Purpose     | Testing / debugging    | Production automation     |
| Input data  | Often sample/test data | Real-time/live data       |
| Repetition  | One-time run           | Continuous or event-based |
| Environment | Development            | Production                |

---

# 4. Important Insight

In real systems:

* You **design workflows using manual execution**
* You **deploy workflows using automatic execution**

Manual execution is for building.
Automatic execution is for running at scale.

---

# 5. Common Mistake

Many beginners assume:

> “If it works manually, it will work automatically”

Not always true because:

* Real triggers send different data formats
* Authentication may differ
* Timing and concurrency matter

---

# 6. Mental Model

Think of it like this:

* Manual execution = **pressing a button to test a machine**
* Automatic execution = **machine running on its own in production**

---


## Error Handling Basics in n8n

In **n8n**, error handling refers to how a workflow behaves when a node fails during execution. Since workflows depend on external systems (APIs, databases, services), failures are normal and must be managed explicitly.

---

# 1. What is an error in n8n?

An error occurs when a node cannot complete its operation.

### Common causes:

* API request fails (404 / 500 / timeout)
* Wrong authentication
* Invalid JSON structure
* Missing required fields
* Database connection failure

---

# 2. Default Behavior (No Error Handling)

By default:

```text id="e1"
Trigger → Node A → Node B (fails) → Workflow stops
```

### Result:

* Execution stops immediately
* Downstream nodes do not run
* Workflow is marked as failed

---

# 3. Basic Error Handling Approaches

## A) Continue On Fail (Node Level)

This is the simplest form.

### What it does:

* Node does not stop workflow if it fails
* Returns error inside output instead

### Behavior:

```text id="e2"
Node A → Node B (fails but continues) → Node C
```

### Output example:

```json id="e3"
{
  "error": "Request failed",
  "status": "failed"
}
```

### Use case:

* Non-critical API calls
* Optional data enrichment

---

## B) IF Node for Error Checking

You manually check if a node failed.

### Example logic:

```text id="e4"
IF error exists?
   → Yes: send alert
   → No: continue workflow
```

### Example expression:

```text id="e5"
{{ $json.error !== undefined }}
```

---

## C) Error Workflow (Global Handling)

n8n allows a dedicated **error workflow**.

### How it works:

* If any workflow fails
* It triggers a separate workflow automatically

### Flow:

```text id="e6"
Main Workflow (fails)
        ↓
Error Workflow triggered
```

### Use case:

* Logging failures
* Sending alerts (Slack/Email)
* Monitoring systems

---

## D) Try-Catch Style (Code Node)

Inside a **Code Node**, you can handle errors manually:

```javascript id="e7"
try {
  const result = JSON.parse($json.data);
  return [{ json: result }];
} catch (error) {
  return [{ json: { error: error.message } }];
}
```

### Use case:

* Complex transformations
* Custom API logic

---

# 4. Retry Mechanism (Important)

Some nodes support retries automatically.

### Example settings:

* Retry 3 times
* Wait 5 seconds between retries

### Flow:

```text id="e8"
API Call → Fail → Retry → Retry → Success / Fail
```

### Use case:

* Unstable APIs
* Network issues

---

# 5. Logging Errors

Best practice is to **store errors instead of ignoring them**.

### Example:

* Save failed records to database
* Send error details to Slack
* Store logs in Google Sheets

---

# 6. Real-World Example

```text id="e9"
Webhook Trigger
      ↓
API Call (OpenAI)
      ↓
IF Success?
   ├── Yes → Send Email
   └── No  → Log Error + Notify Admin
```

---

# 7. Key Principles

1. **Never assume external systems are reliable**
2. Always design fallback paths
3. Separate critical and non-critical steps
4. Log everything important
5. Fail gracefully, not silently

---

# 8. Mental Model

Think of error handling as:

> “What should happen when something breaks?”

Instead of letting workflows crash, you design:

* fallback paths
* retries
* logging
* alerts
